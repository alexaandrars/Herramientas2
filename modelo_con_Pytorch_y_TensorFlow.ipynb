## Reducción del tamaño de las imágenes
Lo primero será, una vez realizada la selección de imágenes, reducir su tamaño. En este caso se va a emplear un tamaño de 28x28 píxeles.

Para acceder a las imágenes se han subido a Google Drive y se ha realizado la conexión de Colab y este para permitir el acceso. Se recomienda hacer lo mismo para que las rutas de acceso funcionen.

Las imágenes se encuentran accesibles en el siguiente enlace:

```python
import cv2
import os

# Rutas de entrada y salida
carpeta_entrada = '/content/drive/My Drive/imagenes/Seleccion de imagenes sin cambios/AK'
carpeta_salida = '/content/drive/My Drive/imagenes/Imagenes tamaño bueno/AK tamaño bueno'

# Crear carpeta de salida si no existe
if not os.path.exists(carpeta_salida):
    os.makedirs(carpeta_salida)

# Iterar sobre las imágenes
for archivo in os.listdir(carpeta_entrada):
    ruta_imagen = os.path.join(carpeta_entrada, archivo)

    if os.path.isfile(ruta_imagen) and archivo.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Cargar la imagen
        imagen = cv2.imread(ruta_imagen)

        # Redimensionar
        imagen_redimensionada = cv2.resize(imagen, (150, 150))

        # Guardar la imagen redimensionada
        ruta_salida = os.path.join(carpeta_salida, f'redimensionada_{archivo}')
        cv2.imwrite(ruta_salida, imagen_redimensionada)

print("Imágenes redimensionadas y guardadas")
```

## Convertir imágenes a escala de grises

```python
import cv2
import os

# Ruta de la carpeta con imágenes originales
carpeta_entrada = '/content/drive/My Drive/imagenes/Imagenes tamaño bueno/AK tamaño bueno'

# Ruta de la carpeta donde guardar las imágenes en blanco y negro
carpeta_salida = '/content/drive/My Drive/imagenes/imagenes finales/AK'

# Crear la carpeta de salida si no existe
if not os.path.exists(carpeta_salida):
    os.makedirs(carpeta_salida)

# Iterar sobre todas las imágenes de la carpeta de entrada
for nombre_archivo in os.listdir(carpeta_entrada):
    # Construir la ruta completa de cada archivo
    ruta_imagen = os.path.join(carpeta_entrada, nombre_archivo)

    # Comprobar que sea un archivo de imagen válido
    if os.path.isfile(ruta_imagen) and nombre_archivo.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        # Cargar la imagen
        imagen_color = cv2.imread(ruta_imagen)

        # Convertir a escala de grises
        imagen_bn = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)

        # Guardar la imagen convertida en la carpeta de salida
        ruta_salida = os.path.join(carpeta_salida, f'bn_{nombre_archivo}')
        cv2.imwrite(ruta_salida, imagen_bn)

print("Todas las imágenes han sido convertidas y guardadas")
```

## Normalización

Por último, antes de empezar con el modelo, se van a normalizar las imágenes de forma que cada píxel tenga un valor entre 0 y 1. Para ello se divide el valor de cada píxel entre 255. Esto facilitará el entrenamiento del modelo.

```python
import cv2
import os
import numpy as np

# Ruta de la carpeta con las imágenes
ruta_carpeta = '/content/drive/My Drive/imagenes/imagenes finales/Acne'

# Ruta de la carpeta de salida donde se guardarán las imágenes normalizadas
ruta_salida = '/content/drive/My Drive/imagenes/imagenes finales normalizadas/Acne'

# Asegúrate de que la carpeta de salida exista, si no, créala
os.makedirs(ruta_salida, exist_ok=True)

# Iterar sobre las imágenes en la carpeta
for nombre_imagen in os.listdir(ruta_carpeta):
    if nombre_imagen.endswith('.jpg') or nombre_imagen.endswith('.png'):  # Filtra los archivos de imagen
        # Cargar la imagen
        imagen = cv2.imread(os.path.join(ruta_carpeta, nombre_imagen))

        # Normalizar la imagen
        imagen_normalizada = imagen / 255.0

        # Convertir la imagen normalizada a rango [0, 255] para guardarla en formato correcto
        imagen_normalizada = (imagen_normalizada * 255).astype(np.uint8)

        # Crear el nombre de archivo para la imagen normalizada
        nombre_salida = os.path.join(ruta_salida, nombre_imagen)

        # Guardar la imagen normalizada
        cv2.imwrite(nombre_salida, imagen_normalizada)

# Confirmación de que las imágenes han sido guardadas
print(f"Se han guardado las imágenes normalizadas en {ruta_salida}.")
```

# Cración del DatSet

```python
import matplotlib.pyplot as plt
import tensorflow as tf

# Ruta de la carpeta raíz donde están organizadas las imágenes por clase
ruta_dataset = '/content/drive/My Drive/imagenes/imagenes finales normalizadas'

# Crear el dataset desde las imágenes (aún no desagrupado)
raw_dataset = tf.keras.utils.image_dataset_from_directory(
    ruta_dataset,
    labels="inferred",       # TensorFlow infiere las etiquetas desde las carpetas
    label_mode="int",        # Etiquetas como enteros
    color_mode="grayscale",  # Las imágenes están en escala de grises
    image_size=(150, 150),
    shuffle=False,           # No mezclar para preservar el orden
    seed=42                  # Semilla para reproducibilidad
)

# Obtener nombres de las clases desde el dataset sin procesar
nombres_clases = raw_dataset.class_names

# Desagrupar las imágenes
dataset = raw_dataset.unbatch()

# Mostrar el mapeo entre nombres de clases y etiquetas
print("Mapeo entre nombres de clases y etiquetas:")
for i, nombre in enumerate(nombres_clases):
    print(f"{i}: {nombre}")

# Contar el número total de imágenes
total_imagenes = sum(1 for _ in dataset)

# Dividir en 80% para entrenamiento y 20% para prueba
train_size = int(0.8 * total_imagenes)
test_size = total_imagenes - train_size

# Mezclar y dividir el dataset
dataset = dataset.shuffle(total_imagenes, seed=42)
datos_entrenamiento = dataset.take(train_size)
datos_pruebas = dataset.skip(train_size)

# Verificar la estructura
print(f"\nTotal de imágenes: {total_imagenes}")
print(f"Imágenes en entrenamiento: {train_size}")
print(f"Imágenes en prueba: {test_size}")

# Mostrar una de las imágenes del conjunto de entrenamiento
for imagen, etiqueta in datos_entrenamiento.take(1):
    plt.imshow(imagen.numpy().squeeze(), cmap="gray")  # Remover la dimensión extra del canal
    plt.title(f"Etiqueta: {etiqueta.numpy()} ({nombres_clases[etiqueta.numpy()]})")
    plt.axis("off")
    plt.show()
```

# Implementación del Modelo

```python
# Crear el modelo
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(150,150,1)),
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(9, activation=tf.nn.softmax),
])
```

```python
#Compilar el modelo
modelo.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
TAMANO_LOTE = 32

datos_entrenamiento = datos_entrenamiento.repeat().shuffle(1425).batch(TAMANO_LOTE)
datos_pruebas = datos_pruebas.batch(TAMANO_LOTE)

import math

#Entrenar
ial = modelo.fit(datos_entrenamiento, epochs=5, steps_per_epoch= math.ceil(1425/TAMANO_LOTE))
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(ial.history["loss"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Obtener las imágenes y etiquetas del conjunto de prueba
for imagenes_prueba, etiquetas_prueba in datos_pruebas.take(1):
    imagenes_prueba = imagenes_prueba.numpy()
    etiquetas_prueba = etiquetas_prueba.numpy()
    predicciones = modelo.predict(imagenes_prueba)

# Función para graficar la imagen y la predicción
def graficar_imagen(i, arr_predicciones, etiquetas_reales, imagenes):
    arr_predicciones, etiqueta_real, img = arr_predicciones[i], etiquetas_reales[i], imagenes[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img[..., 0], cmap=plt.cm.binary)

    etiqueta_prediccion = np.argmax(arr_predicciones)
    if etiqueta_prediccion == etiqueta_real:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel(
        "{} {:2.0f}% ({})".format(
            nombres_clases[etiqueta_prediccion],
            100 * np.max(arr_predicciones),
            nombres_clases[etiqueta_real]
        ),
        color=color
    )

# Función para graficar la barra de predicciones
def graficar_valor_arreglo(i, arr_predicciones, etiqueta_real):
    arr_predicciones, etiqueta_real = arr_predicciones[i], etiqueta_real[i]
    plt.grid(False)
    plt.xticks(range(len(nombres_clases)), nombres_clases, rotation=90)
    plt.yticks([])
    grafica = plt.bar(range(len(nombres_clases)), arr_predicciones, color="#777777")
    plt.ylim([0, 1])
    etiqueta_prediccion = np.argmax(arr_predicciones)

    grafica[etiqueta_prediccion].set_color('red')
    grafica[etiqueta_real].set_color('blue')

# Parámetros de visualización
filas = 5
columnas = 5
num_images = min(filas * columnas, len(imagenes_prueba))

plt.figure(figsize=(2 * 2 * columnas, 2 * filas))
for i in range(num_images):
    plt.subplot(filas, 2 * columnas, 2 * i + 1)
    graficar_imagen(i, predicciones, etiquetas_prueba, imagenes_prueba)
    plt.subplot(filas, 2 * columnas, 2 * i + 2)
    graficar_valor_arreglo(i, predicciones, etiquetas_prueba)

plt.tight_layout()
plt.show()
```

```python
import keras_tuner as kt
from tensorflow.keras.callbacks import EarlyStopping

# Definir la función de creación del modelo
def crear_modelo_tuner(hp):
    modelo = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(150, 150, 1)),
        tf.keras.layers.Dense(hp.Int("neuronas_capa1", min_value=64, max_value=256, step=64), activation='relu'),
        tf.keras.layers.Dropout(hp.Float("dropout_rate", min_value=0.1, max_value=0.5, step=0.1)),
        tf.keras.layers.Dense(hp.Int("neuronas_capa2", min_value=32, max_value=128, step=32), activation='relu'),
        tf.keras.layers.Dropout(hp.Float("dropout_rate", min_value=0.1, max_value=0.5, step=0.1)),
        tf.keras.layers.Dense(9, activation='softmax')
    ])
    modelo.compile(
        optimizer=tf.keras.optimizers.Adam(hp.Choice("learning_rate", [0.01, 0.001, 0.0001])),
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )
    return modelo

# Configurar la búsqueda con Hyperband
tuner = kt.Hyperband(
    crear_modelo_tuner,
    objective="val_accuracy",
    max_epochs=50,  # Incrementamos el máximo para capturar el mejor número de épocas
    factor=3,
    directory="busqueda_hiperparametros",
    project_name="modelo_mejorado"
)

# Callback de parada anticipada
callback_early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=5,  # Número de épocas sin mejora antes de detener el entrenamiento
    restore_best_weights=True
)

# Realizar la búsqueda
tuner.search(
    datos_entrenamiento,
    validation_data=datos_pruebas,
    steps_per_epoch=math.ceil(1425 / TAMANO_LOTE),
    epochs=50,
    callbacks=[callback_early_stopping]
)

# Obtener los mejores hiperparámetros
mejores_hiperparametros = tuner.get_best_hyperparameters(1)[0]
print(f"Mejores hiperparámetros: {mejores_hiperparametros.values}")

# Construir el mejor modelo
mejor_modelo = tuner.hypermodel.build(mejores_hiperparametros)

# Entrenar el mejor modelo nuevamente, rastreando el número de épocas
historial_mejor_modelo = mejor_modelo.fit(
    datos_entrenamiento,
    validation_data=datos_pruebas,
    steps_per_epoch=math.ceil(1425 / TAMANO_LOTE),
    epochs=50,
    callbacks=[callback_early_stopping]
)

# Guardar el número de épocas que resultaron en el mejor modelo
mejores_hiperparametros.values["num_epocas"] = len(historial_mejor_modelo.history['accuracy'])
print(f"Número de épocas: {mejores_hiperparametros.values['num_epocas']}")
```

## Imágenes en escala de grises


```python
import cv2
import os

# Ruta de la carpeta con imágenes originales
carpeta_entrada = '/content/drive/My Drive/imagenes/Todas las imagenes tamaño bueno/AK tamaño bueno'

# Ruta de la carpeta donde guardar las imágenes en blanco y negro
carpeta_salida = '/content/drive/My Drive/imagenes/todas las imagenes finales/AK'

# Crear la carpeta de salida si no existe
if not os.path.exists(carpeta_salida):
    os.makedirs(carpeta_salida)

# Iterar sobre todas las imágenes de la carpeta de entrada
for nombre_archivo in os.listdir(carpeta_entrada):
    # Construir la ruta completa de cada archivo
    ruta_imagen = os.path.join(carpeta_entrada, nombre_archivo)

    # Comprobar que sea un archivo de imagen válido
    if os.path.isfile(ruta_imagen) and nombre_archivo.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        # Cargar la imagen
        imagen_color = cv2.imread(ruta_imagen)

        # Convertir a escala de grises
        imagen_bn = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)

        # Guardar la imagen convertida en la carpeta de salida
        ruta_salida = os.path.join(carpeta_salida, f'bn_{nombre_archivo}')
        cv2.imwrite(ruta_salida, imagen_bn)

print("Todas las imágenes han sido convertidas y guardadas")
```

## Normalización

```python
import cv2
import os
import numpy as np

# Ruta de la carpeta con las imágenes
ruta_carpeta = '/content/drive/My Drive/imagenes/todas las imagenes finales/Acne'

# Ruta de la carpeta de salida donde se guardarán las imágenes normalizadas
ruta_salida = '/content/drive/My Drive/imagenes/todas las imagenes finales normalizadas/Acne'

# Asegúrate de que la carpeta de salida exista, si no, créala
os.makedirs(ruta_salida, exist_ok=True)

# Iterar sobre las imágenes en la carpeta
for nombre_imagen in os.listdir(ruta_carpeta):
    if nombre_imagen.endswith('.jpg') or nombre_imagen.endswith('.png'):  # Filtra los archivos de imagen
        # Cargar la imagen
        imagen = cv2.imread(os.path.join(ruta_carpeta, nombre_imagen))

        # Normalizar la imagen
        imagen_normalizada = imagen / 255.0

        # Convertir la imagen normalizada a rango [0, 255] para guardarla en formato correcto
        imagen_normalizada = (imagen_normalizada * 255).astype(np.uint8)

        # Crear el nombre de archivo para la imagen normalizada
        nombre_salida = os.path.join(ruta_salida, nombre_imagen)

        # Guardar la imagen normalizada
        cv2.imwrite(nombre_salida, imagen_normalizada)

# Confirmación de que las imágenes han sido guardadas
print(f"Se han guardado las imágenes normalizadas en {ruta_salida}.")
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from collections import Counter
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
import keras_tuner as kt
from tensorflow.keras.callbacks import EarlyStopping
import math

# Ruta de la carpeta raíz donde están organizadas las imágenes por clase
ruta_dataset = '/content/drive/My Drive/imagenes/todas las imagenes finales normalizadas'

# Crear el dataset desde las imágenes (aún no desagrupado)
raw_dataset = tf.keras.utils.image_dataset_from_directory(
    ruta_dataset,
    labels="inferred",       # TensorFlow infiere las etiquetas desde las carpetas
    label_mode="int",        # Etiquetas como enteros
    color_mode="grayscale",  # Las imágenes están en escala de grises
    image_size=(150, 150),
    shuffle=False,           # No mezclar para preservar el orden
    seed=42                  # Semilla para reproducibilidad
)

# Obtener nombres de las clases desde el dataset sin procesar
nombres_clases = raw_dataset.class_names

# Desagrupar las imágenes
dataset = raw_dataset.unbatch()

# Mostrar el mapeo entre nombres de clases y etiquetas
print("Mapeo entre nombres de clases y etiquetas:")
for i, nombre in enumerate(nombres_clases):
    print(f"{i}: {nombre}")

# Contar el número total de imágenes
total_imagenes = 22992

# Dividir en 80% para entrenamiento y 20% para prueba
train_size = int(0.8 * total_imagenes)
test_size = total_imagenes - train_size

# Mezclar y dividir el dataset
dataset = dataset.shuffle(total_imagenes, seed=42)
datos_entrenamiento = dataset.take(train_size)
datos_pruebas = dataset.skip(train_size)

# Verificar la estructura
print(f"\nTotal de imágenes: {total_imagenes}")
print(f"Imágenes en entrenamiento: {train_size}")
print(f"Imágenes en prueba: {test_size}")

# Mostrar una de las imágenes del conjunto de entrenamiento
for imagen, etiqueta in datos_entrenamiento.take(1):
    plt.imshow(imagen.numpy().squeeze(), cmap="gray")  # Remover la dimensión extra del canal
    plt.title(f"Etiqueta: {etiqueta.numpy()} ({nombres_clases[etiqueta.numpy()]})")
    plt.axis("off")
    plt.show()
```

```python
# Calcular los pesos de clase para manejar el desbalance de clases
contador_clases = Counter(etiqueta.numpy() for _, etiqueta in dataset)

# Mostrar el conteo por clase
print("\nDistribución de clases:")
for clase, cantidad in contador_clases.items():
    print(f"Clase {clase} ({nombres_clases[clase]}): {cantidad} imágenes")

# Calcular los pesos de clase
pesos_clase = compute_class_weight(
    class_weight="balanced",
    classes=np.array(list(contador_clases.keys())),
    y=np.array([etiqueta.numpy() for _, etiqueta in dataset])
)

# Crear un diccionario de pesos para usar con TensorFlow
pesos_clase_dict = {clase: peso for clase, peso in zip(contador_clases.keys(), pesos_clase)}
print("\nPesos de clase calculados:")
print(pesos_clase_dict)

# Visualizar la distribución de clases
plt.bar(contador_clases.keys(), contador_clases.values())
plt.xticks(ticks=list(contador_clases.keys()), labels=[nombres_clases[c] for c in contador_clases.keys()])
plt.title("Distribución de clases")
plt.xlabel("Clase")
plt.ylabel("Cantidad de imágenes")
plt.show()

# Crear la función de creación del modelo para el tuner
def crear_modelo_tuner(hp):
    modelo = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(150, 150, 1)),
        tf.keras.layers.Dense(hp.Int("neuronas_capa1", min_value=64, max_value=256, step=64), activation='relu'),
        tf.keras.layers.Dropout(hp.Float("dropout_rate", min_value=0.1, max_value=0.5, step=0.1)),
        tf.keras.layers.Dense(hp.Int("neuronas_capa2", min_value=32, max_value=128, step=32), activation='relu'),
        tf.keras.layers.Dropout(hp.Float("dropout_rate", min_value=0.1, max_value=0.5, step=0.1)),
        tf.keras.layers.Dense(len(nombres_clases), activation='softmax')
    ])
    modelo.compile(
        optimizer=tf.keras.optimizers.Adam(hp.Choice("learning_rate", [0.01, 0.001, 0.0001])),
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )
    return modelo

# Configurar la búsqueda con Hyperband
tuner = kt.Hyperband(
    crear_modelo_tuner,
    objective="val_accuracy",
    max_epochs=50,
    factor=3,
    directory="busqueda_hiperparametros",
    project_name="modelo_mejorado"
)

# Callback de parada anticipada
callback_early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    restore_best_weights=True
)

# Realizar la búsqueda
tuner.search(
    datos_entrenamiento.batch(32),  # Batch para que el tuner funcione
    validation_data=datos_pruebas.batch(32),
    steps_per_epoch=math.ceil(train_size / 32),
    epochs=50,
    callbacks=[callback_early_stopping],
    class_weight=pesos_clase_dict  # Pasar los pesos de clase
)

# Obtener los mejores hiperparámetros
mejores_hiperparametros = tuner.get_best_hyperparameters(1)[0]
print(f"Mejores hiperparámetros: {mejores_hiperparametros.values}")

# Construir el mejor modelo
mejor_modelo = tuner.hypermodel.build(mejores_hiperparametros)

# Entrenar el mejor modelo nuevamente
historial_mejor_modelo = mejor_modelo.fit(
    datos_entrenamiento.batch(32),
    validation_data=datos_pruebas.batch(32),
    steps_per_epoch=math.ceil(train_size / 32),
    epochs=50,
    callbacks=[callback_early_stopping],
    class_weight=pesos_clase_dict  # Pasar los pesos de clase
)

# Guardar el número de épocas que resultaron en el mejor modelo
mejores_hiperparametros.values["num_epocas"] = len(historial_mejor_modelo.history['accuracy'])
print(f"Número de épocas: {mejores_hiperparametros.values['num_epocas']}")
```
